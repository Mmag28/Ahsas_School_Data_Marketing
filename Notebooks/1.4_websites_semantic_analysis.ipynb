```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation
    tokens = text.split()
    tokens = [word for word in tokens if word not in ENGLISH_STOP_WORDS and len(word) > 2]
    return tokens

df["tokens"] = df["text"].apply(preprocess_text)
```

```python
school_word_counts = {}
for _, row in df.iterrows():
    school = row["school"]
    tokens = row["tokens"]
    word_freq = Counter(tokens)
    school_word_counts[school] = dict(word_freq.most_common(15))

# Preview
for school, words in school_word_counts.items():
    print(f"{school}: {', '.join(words.keys())}")
```

## üß≠ School Narrative Positioning Analysis  
**Mapping Schools by Website Narrative Alignment on Key Cultural Axes**

---

### üéØ Objective

The goal of this analysis was to position a set of schools based on the semantic content of their **website narratives**. Specifically, we aimed to **map each school along two cultural dimensions**:

- **X-axis**: _Whole-Person Growth_ ‚Üê‚Üí _Academic Prestige_  
- **Y-axis**: _Innovation & Flexibility_ ‚Üê‚Üí _Tradition & Structure_  

These axes were chosen to reflect meaningful contrasts in school identity, philosophy, and educational approach as communicated in their public-facing messaging.

---

### üß† Methodology

1. **Narrative Collection**:  
   We manually gathered the descriptive website content for each school in the dataset.

2. **Preprocessing**:  
   Text was lowercased, cleaned of punctuation, and pre-tokenized. No aggressive lemmatization or stemming was applied to preserve semantic integrity.

3. **Axis Word Banks Definition**:  
   For each end of the two axes, we curated lists of ~25‚Äì30 keywords that semantically represent the pole. These included:
   - Developmental, emotional, and student-centered terms (e.g., `belonging`, `resilience`)
   - Competitive, prestige-oriented academic signals (e.g., `ivy league`, `AP courses`)
   - Creative, flexible, and exploratory language (e.g., `maker`, `project-based`)
   - Traditional/structured language (e.g., `routine`, `rubrics`, `discipline`)

4. **TF-IDF Vectorization**:  
   We used `TfidfVectorizer` from scikit-learn to convert each school‚Äôs narrative into a high-dimensional feature vector based on term importance across the corpus.

5. **Semantic Similarity Scoring**:  
   We computed **cosine similarity** between each school‚Äôs TF-IDF vector and the averaged TF-IDF vectors of each axis pole (e.g., `Whole-Person Growth`, `Academic Prestige`, etc.). These similarity scores became the X and Y coordinates.

6. **Visualization**:  
   Each school was plotted on a 2D chart where:
   - Right side = Greater alignment with _Whole-Person Growth_
   - Left side = Greater alignment with _Academic Prestige_
   - Top = More _Innovative & Flexible_
   - Bottom = More _Structured & Traditional_

- Schools now spread meaningfully across both axes, compared to earlier clustering due to overlapping keyword design and lack of weighting.
- Use of TF-IDF allows **rare, meaningful terms** (e.g., `belonging`, `self-trust`) to influence positioning without being drowned out by frequent but generic terms.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Define updated axis keyword lists
axes = {
    # ‚û§ X-axis: Academic ‚Üî Developmental

    "X_positive": [  # Developmental Orientation (Whole-Child, Humanistic)
        "growth", "wellbeing", "empathy", "belonging", "self-awareness", "self-expression",
        "confidence", "identity", "social-emotional", "emotional development", "relationship-based",
        "personalized", "individual needs", "inner motivation", "resilience", "student voice",
        "connectedness", "self-trust", "curiosity", "self-discovery", "nurturing", "holistic", 
        "exploration", "agency", "meaning-making"
    ],

    "X_negative": [  # Academic Orientation (Achievement, Content Mastery)
        "achievement", "academic excellence", "curriculum standards", "college readiness",
        "content mastery", "grade level", "rigorous instruction", "transcripts", "assessment", 
        "proficiency", "benchmarks", "test preparation", "standardized testing", "report cards",
        "class rank", "honors courses", "college placement", "outcomes", "performance-based", 
        "selective admission", "learning objectives", "scholarly", "advanced coursework"
    ],

    # ‚û§ Y-axis: Progressive ‚Üî Traditional

    "Y_positive": [  # Progressive / Constructivist Pedagogy
        "inquiry-based", "experiential learning", "constructivist", "project-based", 
        "real-world", "student agency", "choice-driven", "interdisciplinary", 
        "collaborative learning", "hands-on", "maker space", "design thinking", 
        "exploratory", "reflective practice", "student-led", "open-ended", "adaptive",
        "flexible grouping", "creative expression", "discussion-based", "learning by doing"
    ],

   "Y_negative": [
    "core knowledge", "teacher-guided instruction", "explicit teaching", "structured learning",
    "study habits", "academic discipline", "sequenced curriculum", "moral education", "character building",
    "classical texts", "liberal arts", "foundational skills", "grammar", "logic", "rhetoric",
    "clarity", "academic tradition", "timeless principles", "civics", "memorization", 
    "literature", "mathematics mastery", "concentration", "academic focus", "intellectual rigor",
    "clear expectations", "consistent standards", "formal assessments", "respect for authority"
]

}

# Load text data
texts = df["text"].fillna("").tolist()
school_names = df["school"].tolist()

# Vectorize with TF-IDF
vectorizer = TfidfVectorizer(ngram_range=(1, 2), stop_words='english')
tfidf_matrix = vectorizer.fit_transform(texts)
feature_names = vectorizer.get_feature_names_out()

# Create TF-IDF DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), index=school_names, columns=feature_names)

# Compute axis vector representations
axis_vectors = {}
for axis, words in axes.items():
    vec = [0] * len(feature_names)
    for word in words:
        if word in feature_names:
            vec[feature_names.tolist().index(word)] = 1
    axis_vectors[axis] = vec

# For each school, compute the contribution of each axis keyword
results = {}
for school in school_names:
    contributions = {}
    tfidf_row = tfidf_df.loc[school]
    for axis, vec in axis_vectors.items():
        axis_word_scores = {
            word: tfidf_row[word]
            for word in axes[axis]
            if word in tfidf_row and tfidf_row[word] > 0
        }
        top_words = sorted(axis_word_scores.items(), key=lambda x: x[1], reverse=True)[:5]
        contributions[axis] = top_words
    results[school] = contributions

# Display example output
for school, axes_words in results.items():
    print(f"\nüéì {school}")
    for axis, words in axes_words.items():
        word_list = ", ".join([f"{w} ({score:.3f})" for w, score in words])
        print(f"  {axis}: {word_list}")
```
